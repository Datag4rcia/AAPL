## Limpeza




```python
#Limpando os símbolos

df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
#Trocando a porcentagem

df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')

df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
#Encontrando a linha que tem o |
df['rating'].value_counts()
```

```python
# Checkando a linha com o símbolo
df.query('rating == "|"')
```

```python
#Retirando o símbolo do rating

df['rating'] = df['rating'].str.replace('|', '3.9').astype('float64')
```

```python
#Trocando o símbolo do rating_count

df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

```python
#Tirando as colunas que não vamos usar

df.drop(['img_link','product_name','about_product','user_name','img_link','product_link'], axis=1, inplace=True)
df.head()
```

## Modelagem estatística
Decidi fazer um modelo de regressão para entender se o preço varia em relação aos outros atributos, assim podemos entender melhor como o mercado onde estes produtos estão situados funciona. Se a precificação e a questão dos descontos  muda de acordo com as avaliações ou se nada disso é importante para a questões já levantadas.

```python
#baixando e imoortando as bibliotecas necessárias
!pip install scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neural_network import MLPRegressor
```

```python
# Onde a 4 coluna é a classe
X = df.drop(df.columns[3], axis=1)  # previsores
y = df.iloc[:, 3]                   # classe
```

```python
#Criação do treinamento e teste
validation_size = 0.2

train_size = int(len(X) * (1-validation_size))
X_train, X_test = X[0:train_size], X[train_size:len(X)]
y_train, y_test = y[0:train_size], y[train_size:len(X)]
```

```python
import pickle
with open('credit.pkl', mode = 'wb') as f:
  pickle.dump([X_train, y_train, X_test, y_test], f)
```

```python
#Validação cruzada
num_folds = 10
seed = 7
scoring = 'neg_mean_squared_error'
```

```python
#aplicando os modelos
models = []
models.append(('LR', LinearRegression()))
models.append(('LASSO', Lasso()))
models.append(('EN', ElasticNet()))
models.append(('KNN', KNeighborsRegressor()))
models.append(('CART', DecisionTreeRegressor()))
models.append(('SVR', SVR()))
```

```python
# Metodo de boosting
models.append(('ABR', AdaBoostRegressor()))
models.append(('GBR', GradientBoostingRegressor()))
# Metodo de bagging
models.append(('RFR', RandomForestRegressor()))
models.append(('ETR', ExtraTreesRegressor()))
```

```python
#Comparando os modelos, aplicando métricas e plotando em um gráfico
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error
import pandas as pd
import matplotlib.pyplot as plt

names = []
kfold_results = []
test_results = []
train_results = []

for name, model in models:
    names.append(name)

    # Criar pipeline com imputação
    pipeline = make_pipeline(SimpleImputer(strategy='mean'), model)

    # K-Fold com shuffle ativado
    kfold = KFold(n_splits=num_folds, shuffle=True, random_state=seed)

    # Cross-validation com tratamento de NaNs
    cv_scores = -1 * cross_val_score(pipeline, X_train, y_train, cv=kfold, scoring=scoring)
    kfold_results.append(cv_scores)

    # Treinamento completo
    pipeline.fit(X_train, y_train)
    train_pred = pipeline.predict(X_train)
    test_pred = pipeline.predict(X_test)

    train_mse = mean_squared_error(y_train, train_pred)
    test_mse = mean_squared_error(y_test, test_pred)

    train_results.append(train_mse)
    test_results.append(test_mse)

    msg = f"{name}: CV MSE = {cv_scores.mean():.4f} (±{cv_scores.std():.4f}), Train MSE = {train_mse:.4f}, Test MSE = {test_mse:.4f}"
    print(msg)

# Criar DataFrame de resultados
results_df = pd.DataFrame({
    'Model': names,
    'CV_MSE_Mean': [res.mean() for res in kfold_results],
    'CV_MSE_Std': [res.std() for res in kfold_results],
    'Train_MSE': train_results,
    'Test_MSE': test_results
})

# Mostrar e plotar os resultados
print("\nResultados ordenados por Test MSE:")
print(results_df.sort_values(by='Test_MSE'))

plt.figure(figsize=(12, 6))
x = range(len(names))

plt.bar(x, results_df['Train_MSE'], width=0.25, label='Train MSE', align='center')
plt.bar([i + 0.25 for i in x], results_df['Test_MSE'], width=0.25, label='Test MSE', align='center')
plt.bar([i + 0.50 for i in x], results_df['CV_MSE_Mean'], width=0.25, label='CV MSE', align='center')

plt.xticks([i + 0.25 for i in x], names, rotation=45)
plt.ylabel('Mean Squared Error')
plt.title('Comparação dos Modelos')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#mostrando o melhor modelo

import seaborn as sns

# Encontrar índice do melhor modelo com menor Test MSE
best_index = results_df['Test_MSE'].idxmin()
best_model_name = results_df.loc[best_index, 'Model']

# Plot
plt.figure(figsize=(10, 6))
colors = ['orange' if name == best_model_name else 'skyblue' for name in results_df['Model']]

sns.barplot(x='Model', y='Test_MSE', data=results_df, palette=colors)

plt.title(f'Melhor Modelo: {best_model_name}', fontsize=14)
plt.ylabel('Test MSE')
plt.xlabel('Modelos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
